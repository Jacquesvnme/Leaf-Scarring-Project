```python
import cv2
import numpy as np

# Load the water hyacinth image
image_path = r"C:\Users\Henko\Downloads\Leaf-Scarring-Project-main\Leaf-Scarring-Project-main\Individual_Code\Henko Holl\Leaf_1.png"
image = cv2.imread(image_path)
if image is None:
    raise ValueError("Image not found or cannot be loaded.")

cv2.imshow("Original", image)

# Copy the image for processing
result = image.copy()
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define HSV color range for brownish scars
lower_brown = np.array([10, 70, 50])
upper_brown = np.array([18, 180, 150])

# Define HSV color range for yellowish centers
lower_yellow = np.array([22, 80, 80])
upper_yellow = np.array([30, 180, 200])

# Create masks for both brown and yellow regions
brown_mask = cv2.inRange(image_hsv, lower_brown, upper_brown)
yellow_mask = cv2.inRange(image_hsv, lower_yellow, upper_yellow)

# Combine the masks
combined_mask = cv2.bitwise_or(brown_mask, yellow_mask)

# Apply morphology to improve mask quality
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
combined_mask = cv2.morphologyEx(combined_mask, cv2.MORPH_CLOSE, kernel)

# Find contours on the combined mask
contours, _ = cv2.findContours(combined_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Initialize variables for calculations
scar_count = 0
total_scar_area = 0
scar_diameters = []

# Calculate leaf area based on mask dimensions
leaf_area = np.sum(combined_mask > 0)

# Loop through each detected contour (scar)
for contour in contours:
    # Minimum enclosing circle for each scar contour
    (x, y), radius = cv2.minEnclosingCircle(contour)
    diameter = 2 * radius  # Diameter calculation
    scar_area = np.pi * (radius ** 2)  # Approximate area of the scar

    # Filter to ensure we're capturing scars, not noise
    if scar_area > 10:  # Minimum area threshold to ignore small noise
        scar_count += 1
        total_scar_area += scar_area
        scar_diameters.append(diameter)
        
        # Draw the detected scar on the result image for visualization
        cv2.circle(result, (int(x), int(y)), int(radius), (0, 0, 255), 2)  # Red circle around each scar

# Calculate percentage damage of the leaf
if leaf_area > 0:
    damage_percentage = (total_scar_area / leaf_area) * 100
else:
    damage_percentage = 0

# Display results
print(f"Total scars identified: {scar_count}")
print(f"Diameters of each scar (in pixels): {scar_diameters}")
print(f"Total damaged area: {total_scar_area:.2f} pixels")
print(f"Leaf surface area: {leaf_area:.2f} pixels")
print(f"Percentage damage on the leaf: {damage_percentage:.2f}%")

# Show images with identified scars
cv2.imshow('Scars Highlighted', result)
cv2.imshow('Combined Mask', combined_mask)

# Save the result image
cv2.imwrite(r"C:\Users\Henko\Downloads\Leaf-Scarring-Project-main\Leaf-Scarring-Project-main\Individual_Code\Henko Holl\Leaf_1_scars.png", result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the water hyacinth scar image
image_path = r"C:\Users\Henko\Downloads\Leaf-Scarring-Project-main\Leaf-Scarring-Project-main\Individual_Code\Henko Holl\Leaf_1.png"
image = cv2.imread(image_path)
cv2.imshow("Original", image)

# Copy the image for masking
result = image.copy()

# Convert image to HSV color space
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define HSV color range for brownish scars on water hyacinth
lower_brown = np.array([10, 50, 20])   # Lower bound for brown hues
upper_brown = np.array([20, 255, 200]) # Upper bound for brown hues

# Create mask based on the defined brown color range
brown_mask = cv2.inRange(image_hsv, lower_brown, upper_brown)

# Apply mask to get only the brown regions
result = cv2.bitwise_and(result, result, mask=brown_mask)

# Find contours in the brown mask
contours, _ = cv2.findContours(brown_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Initialize a list to keep track of merged contours
merged_contours = []

# Distance threshold in pixels (adjust based on your image resolution)
# Assuming a pixel density of around 100 pixels/cm
distance_threshold = 50  # 0.5 cm in pixels (100 pixels/cm * 0.5 cm)

# Process contours to merge close scars
for contour in contours:
    if len(contour) > 0:
        # Calculate the centroid of the current contour
        M = cv2.moments(contour)
        if M["m00"] != 0:
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])
        else:
            continue

        # Check if this centroid is too close to any existing merged contour centroid
        too_close = False
        for merged_contour in merged_contours:
            # Calculate the distance between the centroids
            merged_M = cv2.moments(merged_contour)
            merged_cX = int(merged_M["m10"] / merged_M["m00"])
            merged_cY = int(merged_M["m01"] / merged_M["m00"])

            distance = np.sqrt((cX - merged_cX) ** 2 + (cY - merged_cY) ** 2)
            if distance < distance_threshold:
                too_close = True
                break

        # If it's not too close to any merged contour, add it
        if not too_close:
            merged_contours.append(contour)

# Count the number of merged scars
scar_count = len(merged_contours)

# Show results
print(f"Total scars identified (merged): {scar_count}")
cv2.imshow('Brown Mask', brown_mask)
cv2.imshow('Result with Brown Scars Highlighted', result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
